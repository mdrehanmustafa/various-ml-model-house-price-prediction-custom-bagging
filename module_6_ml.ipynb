```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
df = pd.read_csv("https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv")
df.head()
```

```python
df['bill_length_mm'].fillna(df['bill_length_mm'].mean(), inplace=True)
df['bill_depth_mm'].fillna(df['bill_depth_mm'].mean(), inplace=True)
df['flipper_length_mm'].fillna(df['flipper_length_mm'].mean(), inplace=True)
df['body_mass_g'].fillna(df['body_mass_g'].mean(), inplace=True)
df['island'].fillna(df['island'].mode(), inplace=True)
df['sex'].fillna(df['sex'].mode(), inplace=True)
df.species.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df['species'] = label_encoder.fit_transform(df['species'])
df['island'] = label_encoder.fit_transform(df['island'])
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

label_encoder=LabelEncoder()
df['species'] = label_encoder.fit_transform(df['species'])
df['island'] = label_encoder.fit_transform(df['island'])
df['sex'] = label_encoder.fit_transform(df['sex'])
x=df.drop("species",axis=1)
y=df["species"]
x_test,x_train,y_test,y_train=train_test_split(x,y,train_size=0.8,random_state=1,stratify=y)
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
for kernel in kernels:
    model = SVC(kernel=kernel, degree=3, gamma='scale', C=1.0)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    # Print Accuracy
    print(f"Kernel: {kernel}, Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    model = SVC(kernel=kernel, degree=3, gamma='scale', C=1.0)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    # Print Accuracy
    print(f"Kernel: {kernel}, Accuracy: {accuracy_score(y_test, y_pred):.4f}")
```

```python
#with hyperparameter tuning

param_grid = {
    'C': [0.5, 1, 5, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto', 0.01, 0.1, 1],
    'degree': [2, 3, 4]  # Only for polynomial kernel
}

# Perform Grid Search
grid_search = GridSearchCV(SVC(), param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(x_train, y_train)

# Get best parameters and best model
best_params = grid_search.best_params_
y_pred = grid_search.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)

# Print results
print(f"Best Parameters: {best_params}")
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Load dataset
df = sns.load_dataset("tips")

# Define features and target
x = df.drop("total_bill", axis=1)
y = df["total_bill"]

# Define categorical and numerical columns
cat_col = ["sex", "smoker", "day", "time"]
num_col = ["tip", "size"]

# Define feature engineering pipelines
num_pipeline = Pipeline(steps=[("imputation", SimpleImputer(strategy="median")), ("scaling", StandardScaler())])
cat_pipeline = Pipeline(steps=[("imputation", SimpleImputer(strategy="most_frequent")),
                               ("encoding", OneHotEncoder(handle_unknown="ignore"))])

preprocessor = ColumnTransformer([("num_pipeline", num_pipeline, num_col),
                                  ("cat_pipeline", cat_pipeline, cat_col)])

# Split dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

# Transform features
x_train = preprocessor.fit_transform(x_train)
x_test = preprocessor.transform(x_test)

# Define multiple regression models
models = {
    "LinearRegression": LinearRegression(),
    "DecisionTreeRegressor": DecisionTreeRegressor(),
    "RandomForestRegressor": RandomForestRegressor(),
    "SVR": SVR(),
    "KNeighborsRegressor": KNeighborsRegressor()
}

# Function to evaluate models
def evaluate_models(x_train, x_test, y_train, y_test, models):
    results = {}
    for model_name, model in models.items():
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)
        mae = mean_absolute_error(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        results[model_name] = {"MAE": mae, "MSE": mse, "R2 Score": r2}

    for model_name, metrics in results.items():
        print(f"{model_name}: MAE = {metrics['MAE']:.4f}, MSE = {metrics['MSE']:.4f}, R2 Score = {metrics['R2 Score']:.4f}")

# Evaluate models
evaluate_models(x_train, x_test, y_train, y_test, models)
```

```python
###dbscan

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data  # Features (sepal & petal length & width)

# Standardize the dataset (important for DBSCAN)
X_scaled = StandardScaler().fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)  # Adjust eps & min_samples for better results
labels = dbscan.fit_predict(X_scaled)

# Reduce dimensions using PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot results
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap="viridis", edgecolors="k", s=100)
plt.title("DBSCAN Clustering on Iris Dataset")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label="Cluster Label")
plt.show()

# Print unique clusters
print(f"Unique clusters: {set(labels)}")
```
